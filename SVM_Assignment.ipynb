Randomized search cv

```python
#Here we have 3 hyperparameters hence we will go with randmized searchCV or gridsearchCV, K-fold works only with 1 hyperparameter
```

```python
clf=SVC()                                                                                      #classifier-clf, SVC is our model

param_grid = [{'kernel':['rbf', 'linear'],'gamma':[0.1,0.5],'C':[10,0.1] }]
gsv=RandomizedSearchCV(clf,param_grid, cv=5, verbose=300)                                    #we re storing randomized search in gsv #verbose is no.of lines we want to see in our output

```

```python
gsv.fit(X_train, Y_train)
```

```python
gsv.best_params_, gsv.best_score_   #it gives the best hyperparameters and the accuracy
```

```python
#Let us build our model
clf=SVC(kernel='rbf', gamma=0.1, C=10)
clf.fit(X_train, Y_train)

```

```python
y_pred=clf.predict(X_test)
```

```python
#let us check the accuracy of our model
from sklearn.metrics import accuracy_score, confusion_matrix
acc=accuracy_score(Y_test, y_pred)*100
print("Accuracy:  ", acc)
confusion_matrix(Y_test, y_pred)
```
