###APRIORI ALGORITHM
Association rules with 10% Support and 70% confidence

```python
frequent_itemsets=apriori(book, min_support=0.1, use_colnames=True)
#we are giving min support as 10%
frequent_itemsets

#itemsets with support values
```

```python
rules=association_rules(frequent_itemsets, metric='confidence', min_threshold=0.7)

rules
#confidence less than 0.5 will not be displayed here
```

```python
rules.sort_values('confidence', ascending=False)
# sorted in decending order
```

```python
len(rules)
```

```python
#maximum lift ratio is 2.32
#if lift ratio is > 1  that indiats a rule that is useful in finding consequent item set
#we have 19 association rules with 10% support and 70% condidence
```

```python
#Histogram for support, lift, confidence
rules[['support','confidence','lift']].hist()
plt.show()

#we can see the max support is between 0.10 and 0.13
#max lift ratio is between 1.80 to 1.90
```

```python
plt.scatter(rules.support, rules.confidence)  #scatterplot
plt.show()
```

APRIORI ALGORITHM
Association rules with 10% Support and 70% confidence

```python
frequent=apriori(df, min_support=0.1, use_colnames=True)
frequent.head()
```

```python
len(frequent)
##we have 2259 support
```

```python
rule=association_rules(frequent, metric='confidence', min_threshold=0.7)
rule.head()
```

```python
rule.sort_values('confidence', ascending=False)
```

```python
rule['lift'].agg(['max', 'min'])
```

```python
rule['confidence'].agg(['max', 'min'])
```

```python
#max lift ratio is 7, min is 1.05
#min confidence is 0.75 and max is 1
```

```python
#Histogram for support, lift, confidence
rule[['support','confidence','lift']].hist()
plt.show()
```

```python
#most of the support is at 0.1 to 0.2
#most of the lift ratio is at 7
```

```python
rule[rule.lift>1]
#association rules with 10% support and 70% confidence and lift ratio>1
```

```python
plt.scatter(rule.support, rule.confidence)  #scatterplot
plt.show()
```
