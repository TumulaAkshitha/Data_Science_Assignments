CALCULATE COSINE SIMILARITY BETWEEN USERS

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
#checking user similarity
user_sim=1-pairwise_distances(user_movies_df.values, metric='cosine')
```

```python
user_sim
```

```python
#convert to dataframe
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df
#rows and columns both are users, user 0 is similar to user 4, user 1 is 0.39% similar to user 5
#we can say 0 index user and 4073 index user are 100% similar
#4th index user and 4074 index user are 70% similar
```

```python
#set index and column names to user id
user_sim_df.index=movies_df.userId.unique()
user_sim_df.columns=movies_df.userId.unique()
```

```python
user_sim_df
```

```python
user_sim_df.iloc[0:5, 0:5]
#all diagonals are zero
```

```python
#filling diagonals with zero
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar users
user_sim_df.idxmax(axis=1)[0:10]

#we can recommend the movies seen by 3rd user to 11th user
```
