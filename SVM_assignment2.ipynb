```python
data=pd.read_csv("forestfires.csv")
```

```python
data.head()
```

```python
X_data=data.iloc[:,0:11]
```

```python
X_data.head(2)
```

```python
Y_data=pd.DataFrame(data.iloc[:,30])
Y_data.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder=LabelEncoder()
X_data[['month', 'day']]=X_data[['month', 'day']].apply(LabelEncoder().fit_transform)
X_data.head(2)
```

```python
Y_data[['size_category']]=Y_data[['size_category']].apply(LabelEncoder().fit_transform)
Y_data.head(2)
```

```python
X=X_data.values
Y=Y_data.values
```

Split data into train and test

```python
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

Randomized searchCV

```python
clf=SVC()
param_grid={'kernel': ['linear', 'rbf'], 'gamma':[0.1,0.01,0.1], 'C':[10,0.1,0.001,0.0001]}
gsv=RandomizedSearchCV(clf, param_grid, cv=10, verbose=200)
```

```python
gsv.fit(X_train, Y_train)
```

```python
gsv.best_params_, gsv.best_score_
#it gives the best hyperparameters and the accuracy
```

```python
#Let us build our model with linear kernal and 0.1 gamma and 10 C:
clf=SVC(kernel='linear', gamma=0.1, C=10)
clf.fit(X_train, Y_train)
y_pred=clf.predict(X_test)
```
